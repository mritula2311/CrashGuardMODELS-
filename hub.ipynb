```python
import json
import requests
import cv2
import pytesseract
import numpy as np
from google.colab.patches import cv2_imshow

# === API Setup ===
url = "https://predict.ultralytics.com"
headers = {"x-api-key": "04a7dfcd890bed4b463deeb01d62619fa6ee175034"}
data = {
    "model": "https://hub.ultralytics.com/models/ycBK7Pgcc0fbEldda52P",  # Your uploaded YOLO model
    "imgsz": 640,
    "conf": 0.25,
    "iou": 0.45
}

# === Image Path ===
image_path = "/content/img8.jpeg"
original_image = cv2.imread(image_path)

# === Send Image to API ===
with open(image_path, "rb") as f:
    response = requests.post(url, headers=headers, data=data, files={"file": f})
response.raise_for_status()

# === Parse API JSON Response ===
result = response.json()
detections = result['images'][0]['results']

# === Loop over all detected plates ===
print("\n‚úÖ Number Plate Characters Detected:")
for det in detections:
    box = det['box']
    x1, y1 = int(box['x1']), int(box['y1'])
    x2, y2 = int(box['x2']), int(box['y2'])

    # Crop the detected plate region
    plate_crop = original_image[y1:y2, x1:x2]

    # Preprocess for OCR
    gray = cv2.cvtColor(plate_crop, cv2.COLOR_BGR2GRAY)
    gray = cv2.bilateralFilter(gray, 11, 17, 17)
    _, thresh = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY)

    # Extract characters using Tesseract
    plate_text = pytesseract.image_to_string(thresh, config='--psm 7')
    plate_cleaned = ''.join(filter(str.isalnum, plate_text))

    # Print the extracted plate number
    print(f"üîπ {plate_cleaned}")

    # Draw on the original image
    cv2.rectangle(original_image, (x1, y1), (x2, y2), (0, 255, 0), 2)
    cv2.putText(original_image, plate_cleaned, (x1, y1 - 10),
                cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255, 0, 0), 2)

# === Show Annotated Image in Colab ===
original_image_rgb = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)
cv2_imshow(original_image_rgb)
```

```python
import json
import requests
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
import easyocr
import re # Import the regular expression module for text cleaning

# === Step 0: (Run once in Colab if needed) ===
# These commands install Tesseract (even though we primarily use EasyOCR, sometimes dependencies overlap)
# and the necessary Python libraries.
# !apt update
# !apt install tesseract-ocr -y
# !pip install opencv-python-headless numpy requests easyocr

# === Step 1: API Setup ===
# Replace "YOUR_API_KEY" with your actual Ultralytics Platform API key
# WARNING: Do not commit your API key directly into shared repositories.
# Consider using environment variables or Colab Secrets for secure storage.
url = "https://predict.ultralytics.com"
headers = {"x-api-key": "04a7dfcd890bed4b463deeb01d62619fa6ee175034"} # <<< REPLACE WITH YOUR KEY IF SHARING
data = {
    "model": "https://hub.ultralytics.com/models/ycBK7Pgcc0fbEldda52P", # Ensure this is your license plate model
    "imgsz": 640,
    "conf": 0.25, # Confidence threshold for detections
    "iou": 0.45   # IoU threshold for non-maximum suppression
}

# === Step 2: Load Image ===
# UPDATE this path to the image you want to process
image_path = "/content/img3.jpeg"
print(f"Attempting to load image from: {image_path}")
original_image = cv2.imread(image_path)

# Check if image loaded successfully
if original_image is None:
    print(f"Error: Could not load image from {image_path}. Please check the path.")
    exit() # Exit the script if the image isn't found

# Create a copy of the image to draw annotations on
annotated_image = original_image.copy()
height, width = original_image.shape[:2] # Get image dimensions

print(f"Image loaded successfully. Dimensions: {width}x{height}")

# === Step 3: Send to YOLO Model (Ultralytics Platform) ===
print(f"\nSending image '{image_path}' to YOLO model via API...")
try:
    with open(image_path, "rb") as f:
        # Send the image file in the 'files' parameter and other data in 'data'
        response = requests.post(url, headers=headers, data=data, files={"file": f})

    # Check if the request was successful
    response.raise_for_status()

    # Parse the JSON response
    result = response.json()

    # Safely extract the detections list
    # The structure is usually result -> images (list) -> first image -> results (list of detections)
    detections = result.get('images', [{}])[0].get('results', [])

    if not detections:
        print("‚ùå No license plates detected by the YOLO model.")
        # Show the original image if nothing is detected
        print("\nShowing original image (no detections):")
        cv2_imshow(original_image)
        exit() # Exit the script if no plates were found

except requests.exceptions.RequestException as e:
    print(f"Error during API request: {e}")
    print("Please check your API key, URL, and internet connection.")
    exit()
except KeyError as e:
    print(f"Error parsing API response JSON: Missing expected key {e}")
    print("Response content:", result) # Print the raw response for debugging
    exit()
except Exception as e:
    print(f"An unexpected error occurred during the API call: {e}")
    exit()

print(f"‚úÖ Detected {len(detections)} potential license plates.")

# === Step 4: Process Each Detected Plate with OCR (EasyOCR) ===
# Initialize EasyOCR Reader once outside the loop for performance
print("\nInitializing EasyOCR reader...")
# Specify languages, e.g., ['en'] for English. You can add more if needed.
reader = easyocr.Reader(['en'])
print("EasyOCR reader initialized.")

print("\n--- Detected Number Plates and OCR Results ---")

# Define annotation parameters
color = (0, 255, 0) # Green color for annotation (BGR format for OpenCV)
thickness = 2
font = cv2.FONT_HERSHEY_SIMPLEX
font_scale = 0.7

for i, det in enumerate(detections):
    box = det.get('box', {}) # Use .get() for safer access
    # Ensure coordinates are integers and handle potential missing keys
    x1 = int(box.get('x1', 0))
    y1 = int(box.get('y1', 0))
    x2 = int(box.get('x2', 0))
    y2 = int(box.get('y2', 0))

    # --- Optional: Add Padding to Bounding Box ---
    # Adding a small border can sometimes help OCR, but be careful not to include too much background
    padding = 5 # pixels
    x1_pad = max(0, x1 - padding)
    y1_pad = max(0, y1 - padding)
    x2_pad = min(width, x2 + padding)
    y2_pad = min(height, y2 + padding)

    # Crop the detected plate region from the original image using padded coordinates
    plate_crop = original_image[y1_pad:y2_pad, x1_pad:x2_pad]

    # Check if the crop is valid (has dimensions)
    if plate_crop.shape[0] == 0 or plate_crop.shape[1] == 0:
        print(f"Warning: Plate crop {i+1} at ({x1_pad},{y1_pad})-({x2_pad},{y2_pad}) is empty or invalid. Skipping OCR.")
        detected_text = "INVALID_CROP" # Indicate failure
        cleaned_text = "INVALID_CROP"
    else:
        # --- Perform OCR on the cropped plate using EasyOCR ---
        try:
            # Read text from the cropped image. detail=0 returns just the text strings.
            ocr_results = reader.readtext(plate_crop, detail=0)

            # Join all detected text strings and remove any spaces EasyOCR might add
            detected_text = "".join(ocr_results).replace(" ", "")
            # Convert to uppercase as license plates are typically uppercase
            detected_text = detected_text.upper()

            # --- Optional: Clean up the detected text ---
            # Use regex to remove any leading or trailing non-alphanumeric characters
            # This helps remove noise like ']', '-', '/', etc. from the ends
            cleaned_text = re.sub(r'^[^A-Za-z0-9]+|[^A-Za-z0-9]+$', '', detected_text)

        except Exception as e:
            print(f"Error during OCR for plate {i+1}: {e}")
            detected_text = "OCR_ERROR"
            cleaned_text = "OCR_ERROR" # Indicate failure

        print(f"Plate #{i+1}: Box=({x1},{y1})-({x2},{y2}), Cropped=({x1_pad},{y1_pad})-({x2_pad},{y2_pad})")
        print(f"  Original OCR Result: '{detected_text}'")
        print(f"  Cleaned OCR Result:  '{cleaned_text}'")

        # --- Visual Check: cropped plate (optional, uncomment if needed) ---
        # print(f"Showing cropped plate #{i+1} for visual check:")
        # cv2_imshow(plate_crop)

    # --- Annotate the annotated_image with the original box and the cleaned text ---
    # Draw the bounding box using the original (non-padded) coordinates from YOLO
    cv2.rectangle(annotated_image, (x1, y1), (x2, y2), color, thickness)

    # Determine text position (just above the original bounding box)
    text_x = x1
    text_y = y1 - 10 # 10 pixels above the top of the box

    # Ensure text doesn't go out of bounds at the top
    if text_y < 15: # Adjust threshold based on font size
        text_y = y2 + 20 # Place text below the box if not enough space above

    # Use the cleaned text for annotation if cleaning was applied, otherwise use detected_text
    text_to_annotate = cleaned_text if 'cleaned_text' in locals() and cleaned_text != "INVALID_CROP" else detected_text

    # Add a black background for the text for better visibility
    (text_width, text_height), baseline = cv2.getTextSize(text_to_annotate, font, font_scale, thickness)

    # Adjust background rectangle position based on text position (above or below)
    if y1 - 10 < 15: # Text placed below
        cv2.rectangle(annotated_image, (text_x, text_y - text_height - baseline), (text_x + text_width, text_y), (0, 0, 0), -1)
        cv2.putText(annotated_image, text_to_annotate, (text_x, text_y - baseline), font, font_scale, (255, 255, 255), thickness) # Use white text on black bg
    else: # Text placed above
        cv2.rectangle(annotated_image, (text_x, text_y - text_height - baseline), (text_x + text_width, text_y), (0, 0, 0), -1) # Black rectangle fill
        cv2.putText(annotated_image, text_to_annotate, (text_x, text_y - baseline), font, font_scale, color, thickness) # Green text on black bg

print("----------------------------------------------")

# === Step 5: Show Annotated Full Image ===
print("\nShowing the original image with detected plates and OCR text:")
# cv2_imshow works with BGR images as returned by cv2.imread
cv2_imshow(annotated_image)

# === Optional: Save the annotated image ===
# output_path = "/content/image_annotated.jpg"
# cv2.imwrite(output_path, annotated_image)
# print(f"\nAnnotated image saved to: {output_path}")
```
